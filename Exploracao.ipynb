Equipe de analystics definiram Nome da coluna em json como a Ideal


```python
key_mapping = {'Nome do Item': 'Nome do Produto',
               'Classificação do Produto':'Categoria do Produto',
               'Valor em Reais (R$)': 'Preço do Produto (R$)',
               'Quantidade em Estoque': 'Quantidade em Estoque',
               'Nome da Loja': 'Filial',
               'Data da Venda': 'Data da Venda'}

key_mapping
```

```python
new_dados_csv = []

for old_dict in dados_csv:
    disct_temp = {}
    for old_key, value in old_dict.items():
        disct_temp[key_mapping[old_key]] = value
    new_dados_csv.append(disct_temp)
new_dados_csv[0]
```

```python
nomes_colunas_csv = list(combined_list[-1].keys())
nomes_colunas_csv
```

```python
path_dados_combinado = '../data_processed/dados_combinados.csv'

with open(path_dados_combinado, 'w') as file:
    writer = csv.DictWriter(file, fieldnames=nomes_colunas)
    writer.writeheader()

    for row in combined_list:
        writer.writerow(row)
```

```python
combined_list [0].get('Data da Venda', 'indisponível')
```

```python
dados_combinado_tabela = [nomes_colunas]

for row in combined_list:
    linha = []
    for coluna in nomes_colunas:
        linha.append(row.get(coluna, 'Indisponivel'))
    dados_combinado_tabela.append(linha)
```

```python
path_dados_combinado = '../data_processed/dados_combinado.csv'

with open (path_dados_combinado, 'w') as file:
    writer = csv.writer(file)
    writer.writerows(dados_combinado_tabela)
```
